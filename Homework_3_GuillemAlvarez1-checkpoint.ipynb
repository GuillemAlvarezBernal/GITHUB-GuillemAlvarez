### a)
 

I'm not particularly fond of the periodic boundary conditions, where if a particle moves off the right-hand side of the grid, it reappears on the left, and so forth. This makes it challenging to generate clear plots. Personally, I prefer a scenario where when a particle hits a wall, it is redirected back to its initial zone, as if a force compels them inwards. However, I acknowledge that this wouldn't accurately represent true Brownian motion, so we maintain the original boundaries.

```python
import numpy as np
import matplotlib.pyplot as plt

seed = 222
rng = np.random.default_rng(seed)

#We set only three general variables
N = 10**4    #number of particles
n = 200      #lattice size
T = 800      #time(uploads of positions)

#We define the function with which we will move one particle to the other side of the lattice when they arrive to the limit
def particle_control(positions_x, positions_y):
    #First we look if some particle is at the boudary
    left = np.where(positions_x == -1)
    right = np.where(positions_x == 201)
    top = np.where(positions_y == 201)
    bottom = np.where(positions_y == -1)
    
    #If there is a particle in the edge, we move them
    if len(left[0]) > 0:
        positions_x[left] = 200
    if len(right[0]) > 0:
        positions_x[right] = 0
    if len(top[0]) > 0:
        positions_y[top] = 0
    if len(bottom[0]) > 0:
        positions_y[bottom] = 200
    return positions_x, positions_y


#We will track all the particles with two arrays, one for the x positions and the other for the y positions
#We generate random positions
positions_x = np.random.randint(0,n+1, N)
positions_y = np.random.randint(0,n+1, N)

#We will keep the trajectories of four particles individually and their distance from the initial point
#We could have used a loop where we display the trajectories of a generic number of particles, but for just four, this is 
#more straight forward
trajectory_1_x = np.array([])
trajectory_1_y = np.array([])
distance_1 = np.array([])

trajectory_2_x = np.array([])
trajectory_2_y = np.array([])
distance_2 = np.array([])

trajectory_3_x = np.array([])
trajectory_3_y = np.array([])
distance_3 = np.array([])

trajectory_4_x = np.array([])
trajectory_4_y = np.array([])
distance_4 = np.array([])

#We start the loop where we will upload all the particles positions
for i in range(T):      
    #We set random move to any direction for x and y, 
    #this will cover all the possible movement in each of the eight possibilities
    move_x = np.random.randint(-1,2, len(positions_x))
    move_y = np.random.randint(-1,2, len(positions_y))
    #we change the possitions by summing the previous positions and the movement
    positions_x = move_x + positions_x
    positions_y = move_y + positions_y
    #then we check for edge particles
    positions_x, positions_y = particle_control(positions_x, positions_y)
            
    #finally we follow the particle trajectories, the distance is the modulus of the difference between the initial point and
    #the actual position
    trajectory_1_x = np.append(trajectory_1_x, positions_x[0])
    trajectory_1_y = np.append(trajectory_1_y, positions_y[0])
    dist_1 = np.sqrt((trajectory_1_x[i]-trajectory_1_x[0])**2+(trajectory_1_y[i]-trajectory_1_y[0])**2)
    distance_1 = np.append(distance_1, dist_1)
    
    trajectory_2_x = np.append(trajectory_2_x, positions_x[1])
    trajectory_2_y = np.append(trajectory_2_y, positions_y[1])
    dist_2 = np.sqrt((trajectory_2_x[i]-trajectory_2_x[0])**2+(trajectory_2_y[i]-trajectory_2_y[0])**2)
    distance_2 = np.append(distance_2, dist_2)
    
    trajectory_3_x = np.append(trajectory_3_x, positions_x[2])
    trajectory_3_y = np.append(trajectory_3_y, positions_y[2])
    dist_3 = np.sqrt((trajectory_3_x[i]-trajectory_3_x[0])**2+(trajectory_3_y[i]-trajectory_3_y[0])**2)
    distance_3 = np.append(distance_3, dist_3)
    
    trajectory_4_x = np.append(trajectory_4_x, positions_x[3])
    trajectory_4_y = np.append(trajectory_4_y, positions_y[3])
    dist_4 = np.sqrt((trajectory_4_x[i]-trajectory_4_x[0])**2+(trajectory_4_y[i]-trajectory_4_y[0])**2)
    distance_4 = np.append(distance_4, dist_4)
    
plt.clf()
plt.figure(figsize=(4,4))
plt.plot(positions_x, positions_y , '.', label = 'Particles', ms=0.8)
plt.plot(trajectory_1_x, trajectory_1_y, '-', label = 'Trajectory 1', color = 'k')
plt.plot(trajectory_2_x, trajectory_2_y, '-', label = 'Trajectory 2', color = 'y')
plt.plot(trajectory_3_x, trajectory_3_y, '-', label = 'Trajectory 3', color = 'm')
plt.plot(trajectory_4_x, trajectory_4_y, '-', label = 'Trajectory 4', color = 'c')
plt.ylabel('y positions')
plt.xlabel('x positions')
plt.legend()
plt.show()
```

### b)

This time, the boundary conditions are configured to remove any particle that exits the cell for the sake of simplicity. This adjustment won't impact the general behavior.

The method I use to check if a particle is near a crystal particle involves a step-by-step process. I first identify particles with X or Y positions adjacent to the crystals, and then I examine whether the indexes of the sorted X and Y positions correspond. Although this method, while not the most efficient, has undergone testing and has proven to be sufficiently fast for generating plots in a reasonable timeframe.

```python
import numpy as np
import matplotlib.pyplot as plt

seed = 222
rng = np.random.default_rng(seed)

#With this function we delete every particle that goes out of our canvas
def particle_control(positions_x, positions_y):
    left = np.where(positions_x == -1)
    positions_x = np.delete(positions_x, left[0])
    positions_y = np.delete(positions_y, left[0])    

    right = np.where(positions_x == 201)
    positions_x = np.delete(positions_x, right[0])
    positions_y = np.delete(positions_y, right[0])
    
    top = np.where(positions_y == 201)
    positions_x = np.delete(positions_x, top[0])
    positions_y = np.delete(positions_y, top[0])
    
    bottom = np.where(positions_y == -1)
    positions_x = np.delete(positions_x, bottom[0])
    positions_y = np.delete(positions_y, bottom[0])
    return positions_x, positions_y

#This function moves the postions of the particles that are near a solid to the solid positions array
def solid_maker(positions_x, positions_y, solid_x, solid_y): 
    #for every solid particle
    for i in range(len(solid_x)):
        frozen_x_locations = np.array([])
        frozen_y_locations = np.array([])
        
        #we sort wich fluid particles can be close to the solid
        frozen_x_locations = np.append(frozen_x_locations, np.where(positions_x == solid_x[i])[0])
        frozen_x_locations = np.append(frozen_x_locations, np.where(positions_x == solid_x[i]+1)[0])
        frozen_x_locations = np.append(frozen_x_locations, np.where(positions_x == solid_x[i]-1)[0])
        frozen_y_locations = np.append(frozen_y_locations, np.where(positions_y == solid_y[i])[0])
        frozen_y_locations = np.append(frozen_y_locations, np.where(positions_y == solid_y[i]+1)[0])
        frozen_y_locations = np.append(frozen_y_locations, np.where(positions_y == solid_y[i]-1)[0])
        
        #that potential candidates for x and y directions are actually particles next to the crystal
        intersection = np.intersect1d(frozen_x_locations, frozen_y_locations, return_indices=True)[0]
        
        #we append this new particles to the solid array
        solid_x = np.append(solid_x, positions_x[intersection.astype(int)])
        solid_y = np.append(solid_y, positions_y[intersection.astype(int)])
        
        #and we delete them from the fluid particles array
        positions_x = np.delete(positions_x, intersection.astype(int))
        positions_y = np.delete(positions_y, intersection.astype(int))    
 
    return positions_x, positions_y, solid_x, solid_y
```

$\textit{np.append}$ is relatively slow, but it simplifies my approach to solving this question. Nonetheless, we can achieve an 80% freeze rate for the particles within a reasonable timeframe, typically between 1 to 2 minutes.

```python
N = 10**4    #Number of particles
n = 200      #size of the canvas
condition = 0.2  #condition to reach the 80%
every_plot = 50   #how many steps before ploting
seed_size = 1     #size of the seed

count = 0
time = np.array([])
num_solid = np.array([])

solid_x = np.array([])
solid_y = np.array([])

#we set the seed for the cristal. I tried to make the seed bigger than one particle, but it was too difficult to 
#set the other particles outside that region

for i in range(seed_size):
    for j in range(seed_size):
        solid_x = np.append(solid_x, 100+i)
        solid_y = np.append(solid_y, 100+j)
        
#Set random initial positions for the fluid particles
positions_x = np.random.randint(0,n+1, N)
positions_y = np.random.randint(0,n+1, N)

#We delat the particles that are generated over the crystal
x_forbidden = np.where(positions_x == 100)
y_forbidden = np.where(positions_y == 100)

positions_x = np.delete(positions_x, np.concatenate((x_forbidden[0],y_forbidden[0])))
positions_y = np.delete(positions_y, np.concatenate((x_forbidden[0],y_forbidden[0])))

num_subplot = 0

while len(positions_x)/N > condition:  
    #as before we change the postions
    move_x = np.random.randint(-1,2, len(positions_x))
    move_y = np.random.randint(-1,2, len(positions_y))
    positions_x = move_x + positions_x
    positions_y = move_y + positions_y
    positions_x, positions_y = particle_control(positions_x,positions_y)
    #we select the new solid particles and we change the liquid particles array
    positions_x, positions_y, solid_x, solid_y = solid_maker(positions_x, positions_y, solid_x, solid_y)
    #we keep track of the number of solid particles
    num_solid = np.append(num_solid, len(solid_x))
    count = count + 1
    time = np.append(time, count)
    #we make the plots 
    if count % every_plot == 0:           
        num_subplot = num_subplot +1
        plt.subplot(3,3, num_subplot)
        plt.title('Time = %g' % (count,))
        plt.plot(positions_x, positions_y , '.', ms=0.5)
        plt.plot(solid_x, solid_y , '.', color = 'r', ms=0.8)

plt.tight_layout()
plt.show()
    
    
plt.clf()
plt.figure(figsize=(4,4))
plt.title('Time = %g' % (count,))
plt.plot(positions_x, positions_y , '.', ms=0.5, label = 'Liquid Particles')
plt.plot(solid_x, solid_y , '.', color = 'r', ms=0.8, label = 'Solid Particles')
plt.ylabel('y positions')
plt.xlabel('x positions')
plt.legend()
plt.show()
```

The crystal is growing in an amorphous manner, which aligns with our expectations. It exhibits a slight preference for rounded shapes. In fact, it features a branched internal structure where we can discern several prominent growth directions. In a perfectly crystalline structure, we should be able to identify eight of these directions, as each particle has eight potential growth paths.

The final plot represents the scenario where approximately 80% of the particles have crystallized.

## 2.Ising model

### a) + b)
 
The boundary conditions dictate that the edge spins are connected to their opposites. Since we are representing all the spins within a single array, transitioning from one row to another necessitates adding or subtracting 'n', that it will be the longitude of one edge.

```python
#we define the function to change one of the spins in the lattice, given the lattice and the index of the spin to change
def new_config(lattice, l):
    if lattice[l] == 1:
        lattice[l] = -1
    else:
        lattice[l] = 1
    return lattice

def Boltzman(E,b):
    return np.exp(-E/b)
```

For the initial configuration, we selected a seed that results in an initial magnetization close to zero. This choice stems from our initial uncertainty about the eventual stabilized regime. However, as we will soon discover, this initial setup may not have been the most advantageous starting point.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate

seed = 245
rng = np.random.default_rng(seed)

n=10 #Number of spins
steps = 10**5  #number of steps

#we set which temperatures we are going to use
T = [1,1.5,1.8,2,2.2,2.4,2.6,3,4,5]

#we set the conficuration for the initial lattice
lattice_0 = rng.uniform(0,1,n**2)
lattice_0[np.where(lattice_0 > 0.5)] = 1 
lattice_0[np.where(lattice_0 < 0.5)] = -1 

#we create the arrays where we are going to store the mean and the variance
mean_M = np.zeros(len(T))
var_M = np.zeros(len(T))

for j in range(len(T)):
    #we will do it for each temperature
    #we want to start with the same configuration for every temperature
    lattice = lattice_0.copy()  
    count = 0
    M = np.zeros(steps)
    m = sum(lattice)
    M[0] = m
    #we start making the changes
    for i in range(steps-1):
        #we decide wich spin we are going to change
        l = np.random.randint(0,len(lattice), 1)
        #and which change in the energy will imply, the surronding spins are selected via the sum/substract of n
        dE = 2*lattice[l]*(lattice[l-1]+lattice[l-(n**2-1)]+lattice[l-n]+lattice[l-(n-1)*n])
        #then we decide if we keep the change or not using the boltzman distribution
        u = np.random.rand()
        if u<= Boltzman(dE,T[j]):
            #if we keep the change we reconficure the lattice
            lattice = new_config(lattice, l)
            #and we calcule the magnetization
            M_new = sum(lattice)
            M[i+1] = M_new
            count = count + 1
        else:
            M[i+1] = M[i]
    print(r"Acceptance fraction for k_BT = %g is %g" % (T[j],count/steps))       
    #We make a plot for each temperature
    t = np.arange(count)
    plt.figure(figsize=(4,4))
    plt.plot(t, M[:len(t)] , '-', ms=1)
    plt.title('$k_BT$ = %g, Acceptance ratio = %g' % (T[j],count/steps))
    plt.ylabel(r'$M$')
    plt.xlabel(r'Iterations')
    plt.show()
    #we store the mean and the variance of the magnetization for every temperature
    mean_M[j] = np.mean(abs(M))
    var_M[j] = np.var(M)
    
```

We can discern two distinct regimes within our observations. The first extends from $k_B T = 1$ to approximately $k_B T = 2$. During this phase, the acceptance ratio is exceptionally low, making it challenging to perceive the behavior of magnetization at the lower values. However, we do notice a propensity for alignment in a certain direction.

In the subsequent temperature values spanning from $k_B T = 1.8$ to $k_B T = 2.2$, the acceptance ratio remains relatively low. Nevertheless, we start to witness the system stabilizing around $M = 100$, which aligns with our expectations. With 100 spins, this outcome is consistent because at lower temperatures, magnetization prevails over randomness, and the system tends toward the minimum energy state where all spins align. It is random if the sistem reaches the stabilization around $M = -100$ or $M = 100$, it depens on the direction of the burn-in phase—characterized by the rapid increase in M during the first 2000 steps.

Subsequently, from $k_B T = 2.6$ to $k_B T = 5$, we observe a swift increase in the acceptance ratio, and magnetization begins to oscillate around zero. This behavior stems from the fact that at higher temperatures, spins tend to distribute randomly, approaching a balanced 50/50 distribution, a behavior reminiscent of an infinite-temperature state. In this phase, thermal fluctuations predominate over magnetization. Notably, we do not observe a burn-in phase here, as we initiated the system with magnetization close to zero.

This transition, where magnetization dominance shifts to thermal fluctuations or vice versa, can be interpreted as a phase transition from a ferromagnetic system to a paramagnetic one. We notice this shift occurring around $k_B T = 2$.

## c)

In the plots below, we can distinctly observe the mean magnetization value dropping from 100 to 0 as the system undergoes a transition, as we had anticipated. This transition becomes evident around $k_B T = 2$.

Additionally, we notice the variance, a statistical measure of the dispersion within a data set, increasing during this transition phase. This high variance reveals the expected significant fluctuations, which are characteristic of second-order phase transitions.
